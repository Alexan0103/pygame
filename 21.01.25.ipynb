```python
import pandas as pd

data = pd.read_csv("healthcare_dataset.csv")
data.head()

```

```python
Ress = list(data["Test Results"])
Res_data = data["Test Results"].value_counts()

Ress = Res_data.index
count_Ress = Res_data.values
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

#   
plt.figure(figsize=(10, 6))
plt.bar(Ress, Res_data, color='green', edgecolor='black', label='Age Data')

#     
plt.title(' ')
plt.xlabel(' ')
plt.ylabel('- ')

plt.grid(axis='y', linestyle='--', alpha=0.7)  #    Y
plt.legend()  # 
#  
plt.tight_layout()
plt.show()
```
